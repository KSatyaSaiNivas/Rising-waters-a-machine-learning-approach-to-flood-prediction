```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn import tree, ensemble, neighbors
from sklearn import metrics
import xgboost

from sklearn.preprocessing import StandardScaler
from joblib import dump
```

```python
dataset = pd.read_excel('../Dataset/flood dataset.xlsx')
dataset.head()
```

```python
dataset.isnull().any()
# dataset.info()
# dataset.describe().T
```

```python
x = dataset.iloc[:, 2:7].values
y = dataset.iloc[:, 9].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(
    x, y, test_size=0.25, random_state=10
)
```

```python
sc = StandardScaler()

x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from xgboost import XGBRegressor

dtree = DecisionTreeRegressor()
Rf = RandomForestRegressor()
knn = KNeighborsRegressor()
xgb = XGBRegressor()


```

```python
dtree.fit(x_train, y_train)
Rf.fit(x_train, y_train)
knn.fit(x_train, y_train)
xgb.fit(x_train, y_train)
```

```python
p1 = dtree.predict(x_test)
p2 = Rf.predict(x_test)
p3 = knn.predict(x_test)
p4 = xgb.predict(x_test)
```

```python
from sklearn.metrics import r2_score, mean_squared_error

p1 = dtree.predict(x_test)
p2 = Rf.predict(x_test)
p3 = knn.predict(x_test)
p4 = xgb.predict(x_test)

print("Decision Tree R2:", r2_score(y_test, p1))
print("Random Forest R2:", r2_score(y_test, p2))
print("KNN R2:", r2_score(y_test, p3))
print("XGBoost R2:", r2_score(y_test, p4))

print("Decision Tree MSE:", mean_squared_error(y_test, p1))
print("Random Forest MSE:", mean_squared_error(y_test, p2))
print("KNN MSE:", mean_squared_error(y_test, p3))
print("XGBoost MSE:", mean_squared_error(y_test, p4))
```
